```python
#Wrangling the data
```

```python
#To convert the Vol. column to float
banks.loc[:, "Vol."] = banks["Vol."].str[:-1].astype(float)

#To delete the "-" character in the Date column
dates = banks["Date"]
new_date = []
for date in dates:
    a=date.replace("-", " ")
    new_date.append(a)
banks["Date"] = new_date

#To convert "Change %" column to float
banks.loc[:, "Change %"] = banks["Change %"].str[:-1].astype(float)

#View the modified df
#fbnh
```

```python
#To rename the "Vol." column to "Vol.(M)"
banks = banks.rename({"Vol.":"Vol.(M)"}, axis=1)
banks
```

```python
#Converting the Date column from a string object
#to a pandas datetime object
import datetime 
   
# Function to convert string to datetime 
def convert(date_time): 
    format = '%y %b' # The format 
    datetime_str = datetime.datetime.strptime(date_time, format) 
   
    return datetime_str 
   
# Driver code 
#date_time = '20 Sep'
#print(convert(date_time))
x=[]
date_time = banks["Date"]
for i in date_time:
    a=convert(i)
    x.append(a)
x
```

```python
#Assigning the value of x to the Date column
banks["Date"]=x
banks
```

```python
#Setting the Date column as the index to enable 
#a concise time-series analysis
banks = banks.set_index('Date')
banks.head()
```

```python
#To visualise the aggregate opening an dclosing prices
#for each bank over a period of 5 years
import plotly.graph_objects as go

bank_name = ["FBNH", "UBA", "GTB", "ACCESS", "ZENITH"]

fig = go.Figure()
fig.add_trace(go.Bar(
    x=bank_name,
    y=[fbnh["Price"].mean(),
       uba["Price"].mean(), 
       gtb["Price"].mean(),
       access["Price"].mean(),
       zenith["Price"].mean()],
    name='Opening Price',
    marker_color="midnightblue"
))
fig.add_trace(go.Bar(
    x=bank_name,
    y=[fbnh["Open"].mean(),
      uba["Open"].mean(),
      gtb["Open"].mean(),
      access["Open"].mean(),
      zenith["Open"].mean()],
    name='Closing Price',
    marker_color='lightblue'
))
#To achieve a transparent/white bg
fig.update_layout({
'plot_bgcolor': 'rgba(0, 0, 0, 0)',
'paper_bgcolor': 'rgba(0, 0, 0, 0)',
},
# Here we modify the tickangle of the xaxis, resulting in rotated labels.
barmode='group', xaxis_tickangle=0)
fig.show()
```

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as soup
import numpy as np
import pandas as pd
import csv
```

```python
url = 'https://www.investing.com/equities/guaranty-bnk-historical-data'
req = Request(url , headers={'User-Agent': 'Mozilla/5.0'})
```

```python
webpage = urlopen(req).read()
page_soup = soup(webpage, "html.parser")
```

```python
#Confirm that our page is accessible
page_soup.title
```

```python
#The target data lies in a table, so we need to scrap all tables
table = page_soup.find("table", id = "curr_table")
a=table.tbody.find_all("tr")
for i in a:
    print(i.get_text())
```

```python
#To determine how many table are available on the page
len(page_soup.table)
```

```python
tables = page_soup.table
for table in tables:
    print(table)
```

```python
tables
```

```python
#To save our table in a list
table = page_soup.find("table")

output_rows = []
for table_row in table.findAll('tr'):
    columns = table_row.findAll('td')
    output_row = []
    for column in columns:
        output_row.append(column.text)
    output_rows.append(output_row)
    
```

```python
#View our table as a list
output_rows
```

```python
#Method 1
import csv
with open('gtb_daily.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(output_rows)
    
gtb_daily = pd.read_csv("gtb_daily.csv")
gtb_daily.head(2)
```
